### Importovanje biblioteka

```python
import pymysql
import pandas as pd
```

### Rad sa MySQL bazom

```python
# Podaci za konektovanje sa bazom
host = "localhost"
port = 3306
user = "root"
password = "12345"
database = "video_players_events"

# Konekcija ka bazi
conn = pymysql.connect(host=host, port=port, user=user, passwd=password.encode('utf-8'), db=database)

# Rad sa kursorom
cursor = conn.cursor()
query = "SELECT * FROM video_players_events"
cursor.execute(query)
data = cursor.fetchall()

# Izdvajanje imena kolona
column_names = [name[0] for name in cursor.description]

cursor.close()
conn.close()
```

### Kreiranje dataframe-a

```python
df = pd.DataFrame(data, columns=column_names)
```

## Analiza

```python
df = pd.read_csv("video_players_events.csv")
df
```

```python
df["month"] = pd.to_datetime(df["created_at"]).dt.month
```

### Pregled gledanosti kurseva po mesecima

```python
df.pivot_table(index="course_id", columns="month", values="total_time", aggfunc="sum")
```

### Pregled gledanosti lekcija po mesecima

```python
df.pivot_table(index="lecture_id", columns="month", values="total_time", aggfunc="sum")
```

### Pregled gledanosti videa po mesecima

```python
df.pivot_table(index="video_id", columns="month", values="total_time", aggfunc="sum")
```
